```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load housing data
housing = pd.read_csv("housing-subset.csv")
```

```python
## BOX SUBPLOTS (STATIC)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.boxplot(data=housing, x="ocean_proximity", y="median_house_value", ax=axes[0])
axes[0].set_title("Median House Value by Ocean Proximity")
axes[0].set_xlabel("Ocean Proximity")
axes[0].set_ylabel("Median House Value (USD)")

sns.boxplot(data=housing, x="ocean_proximity", y="median_income", ax=axes[1])
axes[1].set_title("Median Income by Ocean Proximity")
axes[1].set_xlabel("Ocean Proximity")
axes[1].set_ylabel("Median Income (USD)")

plt.tight_layout()
plt.show()
```

```python
## SCATTERPLOT (STATIC)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=housing, x="median_income", y="median_house_value", hue="ocean_proximity", palette="Set1")

plt.title("Median Income & Median House Value by Ocean Proximity")
plt.xlabel("Median Income (USD)")
plt.ylabel("Median House Value (USD)")
plt.legend(title="Ocean Proximity", loc="upper right")

plt.show()
```
